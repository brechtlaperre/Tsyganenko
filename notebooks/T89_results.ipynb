```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.chdir('..')
```

```python
from DA.preprocess import read_and_parse
from DA.domain import get_results
```

```python
grid, B, _, _, _ = read_and_parse('model/T89/refout/OUT00.DAT') # reference output
grid, cor_ext, cor_magn, average, variance = get_results('model/T89/output', [0], [0]) # perturbed average
```

```python
sns.set(style='white', context='paper')
fig, ax = plt.subplots(1,2, figsize=(10,5))

ln = ax[0].streamplot(grid[0], grid[2], B[0], B[2])
ax[0].plot(0,0, 'go', markersize=10)
ax[0].set_title('T89: Reference')
ax[0].set_yticks(np.arange(-30,40,10))
ax[0].set_xticks(np.arange(-40,30,10))
ax[0].set_xlabel(r'$X/R_E$')
ax[0].set_ylabel(r'$Z/R_E$')

ln = ax[1].streamplot(grid[0], grid[2], average[0], average[2])
ax[1].plot(0,0, 'go', markersize=10)
ax[1].set_title('T89: Ensemble average')
ax[1].set_yticks(np.arange(-30,40,10))
ax[1].set_xticks(np.arange(-40,30,10))
ax[1].set_xlabel(r'$X/R_E$')
ax[1].set_ylabel(r'$Z/R_E$')

plt.tight_layout()
plt.savefig('notebooks/T89/T89.eps', format='eps')
plt.savefig('notebooks/T89/T89.png', format='png')
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))
plt.streamplot(grid[0], grid[2], average[0], average[2])
plt.plot(0,0, 'go', markersize=10)
plt.title('Fieldlines of the ensemble average of the T89 model')
plt.xlabel(r'$X/R_E$')
plt.ylabel(r'$Z/R_E$')
plt.savefig('notebooks/T89/ens_av_T89.eps', format='eps')
plt.savefig('notebooks/T89/ens_av_T89.png', format='png')
```
