```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 40

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
from DA.preprocess import read_and_parse
from DA.domain import get_results
```

```python
grid, B, _, _, _ = read_and_parse('model/TA15/refout/OUT00.DAT') # reference output
grid, cor_ext, cor_magn, average, variance = get_results('model/TA15/output', [0], [0]) # perturbed average
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,7))

ln = ax[0].streamplot(grid[0], grid[2], B[0], B[2])
ax[0].plot(0,0, 'go', markersize=10)
ax[0].set_title('TA15: Reference', fontsize=18)
ax[0].set_yticks(np.arange(-30,40,10))
ax[0].set_xticks(np.arange(-40,30,10))
ax[0].set_xlabel(r'$X/R_E$', fontsize=15)
ax[0].set_ylabel(r'$Z/R_E$', fontsize=15)

ln = ax[1].streamplot(grid[0], grid[2], average[0], average[2])
ax[1].plot(0,0, 'go', markersize=10)
ax[1].set_title('TA15: Ensemble average', fontsize=18)
ax[1].set_yticks(np.arange(-30,40,10))
ax[1].set_xticks(np.arange(-40,30,10))
ax[1].set_xlabel(r'$X/R_E$', fontsize=15)
ax[1].set_ylabel(r'$Z/R_E$', fontsize=15)

plt.tight_layout()
plt.savefig('notebooks/TA15/TA15_ref_av.eps', format='eps')
plt.savefig('notebooks/TA15/TA15_ref_av.png', format='png')
```
